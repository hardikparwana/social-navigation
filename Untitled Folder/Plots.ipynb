```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
gains = np.array([0.5, 1.0, 1.5, 2.0, 2.5, 3.0])
unaware = np.array([675, 610, 577, 548, 525, 506])
aware = np.array([606, 570, 544, 526, 511, 498])
```

```python
fig, ax = plt.subplots()
ax.plot(gains, unaware, 'r*-', label='Unaware')
ax.plot(gains, aware, 'g*-', label='Aware')
ax.set_title('Cost vs Human\'s repulsion gain')
ax.set_xlabel('Human Repulsion gain')
ax.set_ylabel('Cost')
ax.legend()
```
